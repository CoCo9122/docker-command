## 矩形波のフーリエ級数展開

関数$f(x)$は以下に定義される
$$f(x) = \left\{
\begin{array}{ll}
1 & (0 \leq x \leq \pi) \\
-1 & (-\pi \leq x < 0)
\end{array}
\right.$$
フーリエ級数展開を行うと$f(x)$は
$$ f(x) = \frac{4}{\pi}\sum_{k=1}^{n} \frac{1}{2n-1} \sin\{(2n-1)x\}$$

```python
def squareWave(n, x):
    y = 0
    for i in range(1, n+1, 1):
        y += 4/np.pi/(2*i-1)*np.sin((2*i-1)*x)
    return y

x = np.linspace(-np.pi, np.pi, 10000)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
for i in [1, 5, 10, 20, 100]:
    ax.plot(x, squareWave(i, x), label="n = {}".format(i))
ax.grid()
ax.legend()
plt.show()
```
